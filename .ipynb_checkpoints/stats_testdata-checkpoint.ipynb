```python
import pandas as pd

from thrift.transport import TSocket, TTransport
from thrift.protocol import TBinaryProtocol
from thrift_interface import TextToSpeechNLPService
from thrift_interface.ttypes import TNlpParam
import matplotlib.pyplot as plt
import numpy as np

host = '10.40.34.32'
port = '8131'
socket = TSocket.TSocket(host=host, port=port)
transport = TTransport.TFramedTransport(socket)
protocol = TBinaryProtocol.TBinaryProtocol(transport)
client = TextToSpeechNLPService.Client(protocol)

transport.open()
```

```python
def convert_standard_text(sentence):
    sentence = sentence.replace(' .', '.').replace(' ,', ',').replace(' !', '!').replace(' ?', '?').replace(' :' , ':').replace(' ;', ';')
    sentence = sentence.replace('.', ' .').replace(',', ' ,').replace('!', ' !').replace('?', ' ?').replace(':' , ' :').replace(';', ' ;')
    
    new_input = TNlpParam()
    new_input.text = sentence
    
    new_input.max_L = 101000
    new_input.requestId = 5555
    new_input.speakerId = 0
    new_input.extra_options = {
        "mode": "platform"
    }
    speech_output = client.normalizeAndSplit(new_input)
    
    sentence = sentence.split(' ')
    cmu_sentence = speech_output.result[0].text.split(' ')
    temp = []
    for i in range(len(cmu_sentence)):
        if cmu_sentence[i].startswith('@'): 
            temp.append('<ENG>' + sentence[i] + '</ENG>')
        else: 
            temp.append(sentence[i])
    return ' '.join(temp)
```

```python
def plot_evaluation(df, id=None, username=None):
    sentences_id = []
    sentence_id_vi = []
    sentence_id_code_switch = []
    total_production = 0
    total_vits = 0
    both_ok = 0
    both_bad = 0
    for i in range(len(df)):
        if id and df['annotator'][i] != id:
            continue           
        check1 = 'version0' in df['label_url_v1'][i] and 'Ver1' in df['tts'][i] 
        check2 = 'version0' in df['label_url_v2'][i] and 'Ver2' in df['tts'][i]
           
        if 'BothOK' in df['tts'][i]:
            both_ok += 1
            sentences_id.append(df['id'][i])
            if int(df['label_url_v1'][i].split('/')[-1].split('_')[0][1:]) <= 99:
                sentence_id_vi.append(df['id'][i])
            else:
                sentence_id_code_switch.append(df['id'][i])
                
            continue
        if 'BothBAD' in df['tts'][i]:
            both_bad += 1
            continue
        
        if check1 or check2:
            total_vits += 1
            sentences_id.append(df['id'][i])
            if int(df['label_url_v1'][i].split('/')[-1].split('_')[0][1:]) <= 99:
                sentence_id_vi.append(df['id'][i])
            else:
                sentence_id_code_switch.append(df['id'][i])
        else:
            total_production += 1
    y = np.array([both_ok, total_production, total_vits, both_bad])
    mylabels = ["bothOK", "Production", "VITS", "bothBAD"]

    if not username:
        username = 'All'
    
    plt.pie(y, labels = mylabels, autopct='%1.0f%%', pctdistance=0.5, labeldistance=1.2)
    plt.title(username + ' - ' + str(total_production + total_vits + both_ok + both_bad) + ' votes')
    plt.show() 
    return total_vits, sentences_id, sentence_id_vi, sentence_id_code_switch
```

```python
total_vits, sentences_id, sentence_id_vi, sentence_id_code_switch = plot_evaluation(df_tantc)
sum2 = 0
sum3 = 0
sum4 = 0
for i in range(len(sentences_id)):
    count = 1
    for j in range(i+1, len(sentences_id)):
        if sentences_id[i] == sentences_id[j]:
            count += 1
    if count >= 4: sum4 += 1
    if count >= 3: sum3 += 1
    if count >= 2: sum2 += 1
        
print(len(sentences_id))
print(len(sentence_id_vi))
print(len(sentence_id_code_switch))
print(sum2)
print(sum3)
print(sum4)
711/856
```
